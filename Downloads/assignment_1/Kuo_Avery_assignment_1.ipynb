## Search


A general search pseudocode is as follows:

```
search:
    Initialize frontier (and add start node)
    Initialize explored (and add start node)
    
    while frontier is not empy:
        
        node <- remove a node from frontier
        
        if node is goal, return
        
        for each child of node:
            add child to frontier (based on some conditions)
            add child to explored
```

Now, this is the general idea. How we deal with the frontier (implementation, queue, priority queue, stack, list, dictionary, and so on) and some of the conditions is going to differentiate whether this leads to breadth-first, depth-first, or A$^*$. 


Note. The frontier basically keeps track of what will be considered next. 


Note. We're looking for a path to goal (Problems 1-3 below) and not just an ability to find it; you'll need some bookkepping code to be able to keep track of how to construct the resulting path.

## breadth_first_search

The purpose of this function is to find a path for the robot from start which is represented by ü§ñ to goal which is represented by ‚ù§Ô∏è using breadth-first search. Breadth-first search explores all the neighboring nodes to the start nodes first. This method of searching guarantees the shortest path to goal is found. Frontier is the structure that holds nodes indicated by 'üü®' that may be explored. visited is the structure that holds the node that has been visited before. final_path is the structure that helps us recreate the path that BFS takes. path starts of as goal, in order to work backwards to find the green path. I make a deep copy of world, called world_dp, so that changes are not made to the original world. The function prints a üü© path from start to goal given a set of allowable moves within the search area, world. It marks all the explored nodes in frontier with üü®. 

* **world**: the search area
* **start**: the starting location of the robot
* **goal**: the goal location in the search area
* **moves**: the allowable moves in the search area


**returns** None (but prints out a nice view of the world and replaces appropriate cells using 'üü®' and 'üü©').

````python
def breadth_first_search(world, start, goal, moves): # Amend argument list and description above as you see fit
    frontier, visited, final_path, path, world_dp = deque([start]), set([start]), {}, goal, deepcopy(world)
    while frontier:
        curr_location = frontier.popleft()
        if curr_location == goal: break
        for move in moves:
            neighbor_location = curr_location[0] + move[0], curr_location[1] + move[1]
            if (0 <= neighbor_location[0] < len(world) and 0 <= neighbor_location[1] < len(world[0]) and neighbor_location not in visited and world_dp[neighbor_location[0]][neighbor_location[1]] != '‚¨õ'):
                visited.add(neighbor_location)
                frontier.append(neighbor_location)
                if world_dp[neighbor_location[0]][neighbor_location[1]] == '‚¨ú': world_dp[neighbor_location[0]][neighbor_location[1]] = 'üü®'
                final_path[neighbor_location] = curr_location
    while path != start:
        if path != goal: world_dp[path[0]][path[1]] = 'üü©'
        path = final_path[path]
        world_dp[start[0]][start[1]], world_dp[goal[0]][goal[1]] = 'ü§ñ', '‚ù§Ô∏è'
    for row in world_dp:
        print("".join(row))
    return 
````

## depth_first_search

The purpose of this function is to find a path for the robot from start which is represented by ü§ñ to goal which is represented by ‚ù§Ô∏è using depth-first search. Depth-first search explores all the nodes connected to the start node first. It is not guaranteed to find the shortest path to goal. Frontier is the structure that holds nodes indicated by 'üü®' that may be explored. visited is the structure that holds the node that has been visited before. final_path is the structure that helps us recreate the path that DFS takes. I make a deep copy of world, called world_dp, so that changes are not made to the original world. The function prints a üü© path from start to goal given a set of allowable moves within the search area, world. It marks all the explored nodes in frontier with üü®. 

* **world**: the search area
* **start**: the starting location of the robot
* **goal**: the goal location in the search area
* **moves**: the allowable moves in the search area


**returns** None (but prints out a nice view of the world and replaces appropriate cells using 'üü®' and 'üü©').

````python
def depth_first_search(world, start, goal, moves): # Amend argument list and description above as you see fit
    frontier, visited, final_path, world_dp = [start], set([start]), {}, deepcopy(world)
    while frontier:
        curr_location = frontier.pop()
        visited.add(curr_location)
        if curr_location == goal:
            while curr_location != start:
                if curr_location != goal: world_dp[curr_location[0]][curr_location[1]] = 'üü©'
                curr_location = final_path[curr_location]
            world_dp[start[0]][start[1]], world_dp[goal[0]][goal[1]] = 'ü§ñ', '‚ù§Ô∏è'
            for row in world_dp: 
                print("".join(row))
            return
        for move in reversed(moves):
            neighbor_location = curr_location[0] + move[0], curr_location[1] + move[1]
            if (0 <= neighbor_location[0] < len(world_dp) and 0 <= neighbor_location[1] < len(world_dp[0]) and world_dp[neighbor_location[0]][neighbor_location[1]] != '‚¨õ' and neighbor_location not in frontier and neighbor_location not in visited):
                frontier.append(neighbor_location)
                world_dp[neighbor_location[0]][neighbor_location[1]] = 'üü®'
                final_path[neighbor_location] = curr_location
````

## a_star_search
The purpose of this function is to find a path for the robot from start which is represented by ü§ñ to goal which is represented by ‚ù§Ô∏è using a* search. A* search explores all the nodes using weighted graphs. It uses a heuristic, which is straight line distance in the case of this function. The search aims to find the path from the start to the goal with the smallest cost. Frontier is the structure that holds nodes indicated by 'üü®' that may be explored and the heuristic/cost at that node. final_path is the structure that helps us recreate the path that DFS takes. path is initializes as goal so I can use it as a pointer to backtrack to find the final_path. cost keeps track of the cost at each node in the path. I make a deep copy of world, called world_dp, so that changes are not made to the original world. The function prints a üü© path from start to goal given a set of allowable moves within the search area, world. It marks all the explored nodes in frontier with üü®. 

* **world**: the search area
* **start**: the starting location of the robot
* **goal**: the goal location in the search area
* **moves**: the allowable moves in the search area


**returns** None (but prints out a nice view of the world and replaces appropriate cells using 'üü®' and 'üü©').


````python
def a_star_search(world, start, goal, moves): # Amend argument list and description above as you see fit)
    frontier, final_path, path, cost, world_dp = [(math.sqrt((start[0] - goal[0]) ** 2 + (start[1]- goal[1])**2), start)], {}, goal, {}, deepcopy(world)
    while frontier:
        curr_cost, curr_location = heapq.heappop(frontier)
        for move in moves:
            neighbor_location = curr_location[0] + move[0], curr_location[1] + move[1]
            if (0 <= neighbor_location[0] < len(world) and 0 <= neighbor_location[1] < len(world[0]) and world_dp[neighbor_location[0]][neighbor_location[1]] != '‚¨õ') and (neighbor_location not in cost or curr_cost + 1 < cost[neighbor_location]):
                cost[neighbor_location] = curr_cost + 1
                heapq.heappush(frontier, (curr_cost + 1 + math.sqrt((neighbor_location[0] - goal[0]) ** 2 + (neighbor_location[1] - goal[1]) ** 2), neighbor_location))
                if world_dp[neighbor_location[0]][neighbor_location[1]] == '‚¨ú': world_dp[neighbor_location[0]][neighbor_location[1]] = 'üü®'
                final_path[neighbor_location] = curr_location
    while path != start:
            if path != goal: world_dp[path[0]][path[1]] = 'üü©'
            path = final_path[path]
            world_dp[start[0]][start[1]], world_dp[goal[0]][goal[1]] = 'ü§ñ', '‚ù§Ô∏è'
    for row in world_dp:
            print("".join(row))
    return 
````
